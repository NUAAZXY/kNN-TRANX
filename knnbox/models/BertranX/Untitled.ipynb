```python
data = [[9.13, 20.41], [5.78,2.83], [-0.91, -1.73], [1.1, 6.47]]

bars = ('Preprocessing', 'BERT', 'Grammar', 'PointerNet')

barWidth = 0.4
y1 = [9.13, 5.78, -0.91, 1.1]
y2 = [20.41, 2.83, -1.73, 6.47]
r1 = range(len(y1))
r2 = [x + barWidth for x in r1]

plt.bar(r1, y1, width = barWidth, color = ['red' for i in y1],
           edgecolor = ['black' for i in y1], linewidth = 1)
plt.bar(r2, y2, width = barWidth, color = ['blue' for i in y1],
           edgecolor = ['black' for i in y1], linewidth = 1)
plt.xticks([r + barWidth / 2 for r in range(len(y1))], ['Preprocessing', 'BERT', 'Grammar', 'PointerNet'])
plt.xticks(fontsize=13)
plt.yticks(fontsize=11)


labels = ['CoNaLa', 'Django']
plt.legend(labels, fontsize=13)

plt.ylabel('BLEU', fontsize=13)

plt.show()

plt.rcParams["figure.figsize"] = (8,6)
```
